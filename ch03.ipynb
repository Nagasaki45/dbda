## Exercise 3 - Simple plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 100)
y = x ** 3

plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y = x ^ 3$')
plt.title('Qubic function');
```
